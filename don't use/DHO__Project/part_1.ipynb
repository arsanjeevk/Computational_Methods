```python
#importing required libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#Defining RK4 function
def rk4(f, x0, y0, h, n):
    x = x0
    y = np.array(y0)

    x_values = [x0]
    y_values = [y0]

    for i in range(n):
        k1 = f(x, y)
        k2 = f(x + h/2, y + (h/2)*k1)
        k3 = f(x + h/2, y + (h/2)*k2)
        k4 = f(x + h, y + h*k3)

        y = y + (h/6)*(k1 + 2*k2 + 2*k3 + k4)
        x = x + h

        x_values.append(x)
        y_values.append(y.copy())

    return np.array(x_values), np.array(y_values)

#parameters
m = 1
k = 20
y0 = [1.0, 0.0]
t0 = 0.0
h  = 0.1
n  = 200

#Column labels
columns = [
    "time", "position", "velocity", "acceleration",
    "KE", "PE", "Total_E",
    "F_spring", "F_damping", "F_total",
    "x_exact", "v_exact", "a_exact",
    "error_x", "error_v", "error_a"
]

#Storage dictionary
results = {}

#Damping cases
cases = {
    "Underdamped": 1,
    "Critically Damped": 2*np.sqrt(m*k),
    "Overdamped": 12
}

#main loop for the calculation and graphs of three conditions underdamped, critically damped and overdamped
for case_name, c in cases.items():
    
    #ODE function
    def f(t, y):
        y1, y2 = y
        dy1 = y2
        dy2 = -(c/m)*y2 - (k/m)*y1
        return np.array([dy1, dy2])

    t, y = rk4(f, t0, y0, h, n)

    position = y[:, 0]
    velocity = y[:, 1]
    acceleration = -(c/m)*velocity - (k/m)*position

    KE = 0.5 * m * velocity**2
    PE = 0.5 * k * position**2
    Total_E = KE + PE

    F_damping = -c * velocity
    F_spring = -k * position
    F_total = m * acceleration

    #Analytical solutions
    if case_name == "Underdamped":
        omega_d = np.sqrt(4*m*k - c**2)/(2*m)
        gamma = c/(2*m)

        def x_analytical(t):
            return np.exp(-gamma*t) * (
                np.cos(omega_d*t) + (gamma/omega_d)*np.sin(omega_d*t)
            )

        def v_analytical(t):
            return -np.exp(-gamma*t) * (
                omega_d*np.sin(omega_d*t)
                + (gamma**2/omega_d)*np.sin(omega_d*t)
    )

        def a_analytical(t):
            return -(c/m)*v_analytical(t) - (k/m)*x_analytical(t)

    elif case_name == "Critically Damped":
        gamma = c/(2*m)

        def x_analytical(t):
            return (1 + gamma*t) * np.exp(-gamma*t)

        def v_analytical(t):
            return np.exp(-gamma*t) * (gamma - gamma*(1 + gamma*t))

        def a_analytical(t):
            return -(c/m)*v_analytical(t) - (k/m)*x_analytical(t)

    else: #overdamped
        disc = np.sqrt(c**2 - 4*m*k)
        r1 = (-c + disc)/(2*m)
        r2 = (-c - disc)/(2*m)

        A = r2/(r2 - r1)
        B = -r1/(r2 - r1)

        def x_analytical(t):
            return A*np.exp(r1*t) + B*np.exp(r2*t)

        def v_analytical(t):
            return A*r1*np.exp(r1*t) + B*r2*np.exp(r2*t)

        def a_analytical(t):
            return A*(r1**2)*np.exp(r1*t) + B*(r2**2)*np.exp(r2*t)

    x_exact = x_analytical(t)
    error = np.abs(position - x_exact)
    v_exact = v_analytical(t)
    a_exact = a_analytical(t)
    error_v = np.abs(velocity - v_exact)
    error_a = np.abs(acceleration - a_exact)

    #store matrix
    data_matrix = np.column_stack([
        t, position, velocity, acceleration,
        KE, PE, Total_E,
        F_spring, F_damping, F_total,
        x_exact, v_exact, a_exact,
        error, error_v, error_a
    ])

    results[case_name] = data_matrix

    #save CSV
    df = pd.DataFrame(data_matrix, columns=columns)
    df.to_csv(f"{case_name}_matrix.csv", index=False)

    print(f"\n{case_name} Data Preview:")
    print(df.head())
    # print(f"{case_name} Max Error:", np.max(error))

    #plotting
    fig, ax = plt.subplots(2, 3, figsize=(20, 10))
    fig.suptitle(f"{case_name} Harmonic Oscillator", fontsize=16)

    #Displacement
    ax[0,0].plot(t, position, label="RK4")
    ax[0,0].plot(t, x_exact, '--', label="Analytical")
    ax[0,0].set_xlabel("Time")
    ax[0,0].set_ylabel("Position")
    ax[0,0].set_title("Displacement")
    ax[0,0].legend()
    ax[0,0].grid()

    #Velocity
    ax[0,1].plot(t, velocity, label="RK4")
    ax[0,1].plot(t, v_exact, '--', label="Analytical")
    ax[0,1].set_xlabel("Time")
    ax[0,1].set_ylabel("Velocity")
    ax[0,1].set_title("Velocity")
    ax[0,1].legend()
    ax[0,1].grid()

    #Acceleration
    ax[0,2].plot(t, acceleration, label="RK4")
    ax[0,2].plot(t, a_exact, '--', label="Analytical")
    ax[0,2].set_xlabel("Time")
    ax[0,2].set_ylabel("Acceleration")
    ax[0,2].set_title("Acceleration")
    ax[0,2].legend()
    ax[0,2].grid()

    #Energy
    ax[1,0].plot(t, KE, label="KE")
    ax[1,0].plot(t, PE, label="PE")
    ax[1,0].plot(t, Total_E, label="Total")
    ax[1,0].set_xlabel("Time")
    ax[1,0].set_ylabel("Energy")
    ax[1,0].set_title("Energy Decay")
    ax[1,0].legend()
    ax[1,0].grid()

    #Forces
    ax[1,1].plot(t, F_spring, '--', label="Spring")
    ax[1,1].plot(t, F_damping, label="Damping")
    ax[1,1].plot(t, F_total, label="Total")
    ax[1,1].set_xlabel("Time")
    ax[1,1].set_ylabel("Force")
    ax[1,1].set_title("Forces")
    ax[1,1].legend()
    ax[1,1].grid()

    #Phase space
    ax[1,2].plot(position, velocity)
    ax[1,2].set_xlabel("Position")
    ax[1,2].set_ylabel("Velocity")
    ax[1,2].set_title("Phase Space")
    ax[1,2].grid()
    plt.tight_layout()
    plt.show()
```

```python
#Combined position plot
plt.figure()
for case in results:
    plt.plot(results[case][:,0], results[case][:,1], label=case)

plt.xlabel("Time")
plt.ylabel("Position")
plt.title("Comparison of Damping Cases")
plt.legend()
plt.grid()
plt.show()
```
