#Discrete Fourier Transform

$f(t)$ = $\sin t$

and

$t$ = $[0, 2π]$

```python
import numpy as np
import matplotlib.pyplot as plt

N = 1000
L = 2 * np.pi

#formulae
delta_t = L / N
w_max = (2 * np.pi / delta_t) / 2
dw = 2 * np.pi / L

t = np.linspace(0, L, N)
f = np.sin(t)

ft = np.fft.fft(f)

freq = np.fft.fftfreq(N, d=delta_t)
w = 2 * np.pi * freq

fig, ax = plt.subplots(1,2, figsize=(15,5))

#time domain
ax[0].scatter(t, f)
ax[0].set_xlabel("t")
ax[0].set_title("Time domain")

#frequency domain
ax[1].plot(w, np.abs(ft))
ax[1].set_xlabel("ω")
ax[1].set_title("Frequency domain")

plt.tight_layout()
plt.show()
```

```python
print("Wmax =", w_max)
print("dw =", dw)
```
