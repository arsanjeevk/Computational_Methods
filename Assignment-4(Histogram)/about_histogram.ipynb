```python
import numpy as np
import matplotlib.pyplot as plt

#Generate Gaussian data

N = 5000
mu = 0.0          # mean
sigma = 1.0       # standard deviation

data = np.random.normal(mu, sigma, N)   #it will generate the random 5000 numbers with mean(mu) = 0 and standered deviatiation(sigma(std)) = 1

xmin = np.min(data)
xmax = np.max(data)

#Define number of bins
nbins = 50

bin_width = (xmax - xmin) / nbins

# Bin edges
bins = np.linspace(xmin, xmax, nbins + 1)

# Initialize bin counts
counts = np.zeros(nbins, dtype=int)


#Dump data into bins
for x in data:
    # Find bin index
    index = int((x - xmin) / bin_width)

    # Edge case: if x == xmax
    if index == nbins:
        index = nbins - 1

    counts[index] += 1

#Compute histogram; Using numpy default function
# hist, edges = np.histogram(data, bins=bin_edges)

# Plot histogram
plt.figure()
plt.bar(bins[:-1], counts, width=bin_width, align='edge')
plt.xlabel('x')
plt.ylabel('Counts')
plt.title('Histogram (manual binning)')
plt.show()
```

```python
#compare Histogram(manula biggining) and defult inbuilt histogram plot
#using subplot
fig, ax = plt.subplots(1, 2, figsize=(15,5))   # 1 row, 2 columns

#Graph 1
ax[0].bar(bins[:-1], counts, width=bin_width, align='edge')
ax[0].set_xlabel('x')
ax[0].set_ylabel('counts')
ax[0].set_title('Histogram (manual binning)')
ax[0].grid()

#Graph 2
ax[1].hist(data, bins=50)
ax[1].set_xlabel('x')
ax[1].set_ylabel('counts')
ax[1].set_title('using histogram(default)')
ax[1].grid()

plt.tight_layout()
plt.show()
```
