#Question 1

1) Consider the function $f(x) = 2 \cos x − x$. Consider the three initial guesses $x_0 = 0.5$, $π$ &  $2π$ and for
these guesses find the roots using Newton-Raphson method.

```python
#imports
import numpy as np
import matplotlib.pyplot as plt
```

```python
#defining the function
def f(x):
  return 2 * np.cos(x) - x

#derivative of the function
def df(x):
  return -2 * np.sin(x) - 1


x_0 = 0.5  #initial guess
N_max = 100   #maximum number of iteration
tol = 1e-6
roots = []


for i in range(N_max):
    x = x_0 - f(x_0)/df(x_0)
    roots.append(x)

    if abs(x - x_0) < tol and abs(f(x)) < tol:
        break

    x_0 = x

roots = np.array(roots, dtype=float)

#print the results
print("Root =", roots[-1])
print("No of Iterations =", i+1)
print("Iterations:", roots)

```

```python
#Graphs
x = np.linspace(0, 2 * np.pi, 100)

#plot of the function
plt.plot(x, f(x), label='f(x)=2cosx−x')
plt.axhline(0, color='black', linestyle='--')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Plot of the function')
plt.grid()
plt.legend()
plt.show()
```

#Question 2

2) Compare the convergence against bisection method where the ranges are $[0.5, -2]$ , $[0,π]$  and  $[0,2π]$.
Briefly report your observation for both the root finding methods.

```python
#calculating roots using bisection method

#Define Function
def f(x):
  return 2 * np.cos(x) - x

print(f"The value of function at x = 0.5 is : {f(0.5)}")
print(f"The value of function at x = -2 is : {f(-2)}")
print(f"The value of function at x = 0 is : {f(0)}")
print(f"The value of function at x = pi is : {f(np.pi)}")
print(f"The value of function at x = 2pi is : {f(2 * np.pi)}")
print(f"Is the product of the function at x=0.5 and -2 less than zero: {f(0.5) * f(-2) < 0 }")
```

we can see that the value of the function at x = 0 and x = -2 are positive so $f(-2)f(0.5) > 0$ so root does not lies between this range. we can't calculate root for this range using bisection method. for the other given two ranges $f(a)f(b) < 0$ so we can find roots using bisection method

```python
#Take values of x in the range[0, pi]

#finding the roots of f(x) = 0 using bisection method
eps = 1e-3
Nmax = 100  #Maximum number of iterations
a = 0
b = np.pi

#Define Bisection Method
def Bisection(Xminus, Xplus, Nmax, eps):
    roots = []

    for it in range(Nmax):
        x = (Xplus + Xminus) / 2.0
        roots.append(x)

        if f(Xplus) * f(x) > 0.0:
            Xplus = x
        else:
            Xminus = x

        if abs(f(x)) < eps:
            print("Converged in", it+1, "iterations")
            break

    return np.array(roots)
roots = Bisection(a, b, Nmax, eps)

plt.figure()
plt.plot(range(len(roots)), roots, marker='o')
plt.xlabel("Iteration")
plt.ylabel("Root estimate")
plt.title("Convergence of Bisection Method")
plt.grid(True)
plt.show()

print("Root =", roots[-1])
print("Iterations:", roots)
```

```python
#Take values of x in the range[0, 2pi]

#finding the roots of f(x) = 0 using bisection method
eps = 1e-3
Nmax = 100  #Maximum number of iterations
a = 0
b = 2 * np.pi

#Define Bisection Method
def Bisection(Xminus, Xplus, Nmax, eps):
    roots = []

    for it in range(Nmax):
        x = (Xplus + Xminus) / 2.0
        roots.append(x)

        if f(Xplus) * f(x) > 0.0:
            Xplus = x
        else:
            Xminus = x

        if abs(f(x)) < eps:
            print("Converged in", it+1, "iterations")
            break

    return np.array(roots)
roots = Bisection(a, b, Nmax, eps)

plt.figure()
plt.plot(range(len(roots)), roots, marker='o')
plt.xlabel("Iteration")
plt.ylabel("Root estimate")
plt.title("Convergence of Bisection Method")
plt.grid(True)
plt.show()

print("Root =", roots[-1])
print("Iterations:", roots)
```

#Conclusion


The root of $f(x)=2cosx−x$ was computed using both Newton–Raphson and Bisection methods. Newton–Raphson converged much faster and required fewer iterations when the initial guess was close to the root, showing quadratic convergence. However, its performance depended strongly on the starting value. The Bisection method converged more slowly but provided stable and guaranteed convergence when the root was properly bracketed. Thus, Newton–Raphson is faster but sensitive, while Bisection is slower but more reliable.
