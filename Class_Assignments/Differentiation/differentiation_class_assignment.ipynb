$f(x) = sin(x)$

$[a,b] ∈ [0, 2π]$

n = 8, 16, 32, 64


```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#finite difference method to calculate the differentiation
def f(x):
  return np.sin(x)

#for the n=8
n = 8

a = 0
b = 2 * np.pi
h = (b - a )/ n
x = np.linspace(a, b, n+1)

#derivatives
first_forward = (f(x+h) - f(x))/h
first_backward = (f(x) - f(x-h))/h
f_second = (f(x+h) - 2 * f(x) + f(x-h))/ (h ** 2)
f_fourth = (f(x+2*h) - 4*f(x+h) + 6*f(x) - 4*f(x-h) + f(x-2*h)) / (h**4)



plt.plot(first_forward, label="first_forward")
plt.plot(first_backward, label="first_backward")
plt.plot(f_second, label="f_second")
plt.plot(f_fourth, label="f_fourth")
plt.legend()
plt.show()


#Error analysis
df = np.cos(x)
df_second = - np.sin(x)
df_fourth = np.sin(x)

error_first_forward = np.abs(first_forward - df)
error_first_backward = np.abs(first_backward - df)
error_f_second = np.abs(f_second - df_second)
error_f_fourth = np.abs(f_fourth - df_fourth)

print(len(error_first_backward))
x_axis = n+1

#error plot
plt.plot(error_first_forward, label="first_forward")
plt.plot(error_first_backward,  label="first_backward")
plt.plot(error_f_second, label="f_second")
plt.plot(error_f_fourth, label="f fourth")
plt.title("Error Analysis")
plt.legend()
plt.show()
```
