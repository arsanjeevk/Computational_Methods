```python
import numpy as np
import matplotlib.pyplot as plt

initial_data = np.loadtxt('data_with_errors.txt')

reshaped_data = initial_data.reshape(30, 3)

x = reshaped_data[:, 0]
y = reshaped_data[:, 1]
err = reshaped_data[:, 2]

# Weights
w = 1 / err**2

# Required sums
S   = np.sum(w)
Sx  = np.sum(w * x)
Sy  = np.sum(w * y)
Sxx = np.sum(w * x**2)
Sxy = np.sum(w * x * y)

# Least squares formulas
m = (S * Sxy - Sx * Sy) / (S * Sxx - Sx**2)
c = (Sy - m * Sx) / S

print(f"Slope (m) = {m}")
print(f"Intercept (c) = {c}")

# Fitted line
y_fit = m * x + c

# Plot with error bars
plt.errorbar(x, y, yerr=err, fmt='o', label="Experimental data")
plt.plot(x, y, label="Experimental data")
plt.plot(x, y_fit, 'r-', label="Weighted least squares fit")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()

plt.show()
```

#Assignment (Bisection Method)

1) Consider the function $ f(x) = 2 cos x − x$. Consider the range [0-7] and for this range find the roots
using bisection method.


2) Modify the function $f(x) = 2 cos x − x + 5$. Now it has more than one root in the range [0-7]. Please
find all the roots using bisection method.

```python
import numpy as np
import matplotlib.pyplot as plt

#Define Function
def f(x):
  return 2 * np.cos(x) - x

#Take values of x in the range[0, 7]
x = np.linspace(0, 7, 100)
y = f(x)

#plot the function
plt.plot(x, y)
plt.axhline()
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.legend(['f(x) = 2cos(x) - x'])
plt.show()

#finding the roots of f(x) = 0 using bisection method
eps = 1e-3
Nmax = 100  #Maximum number of iterations
a = 0.0
b = 7.0

#Define Bisection Method
def Bisection(Xminus, Xplus, Nmax, eps):
    roots = []

    for it in range(Nmax):
        x = (Xplus + Xminus) / 2.0
        roots.append(x)

        if f(Xplus) * f(x) > 0.0:
            Xplus = x
        else:
            Xminus = x

        if abs(f(x)) < eps:
            print("Converged in", it+1, "iterations")
            break

    return np.array(roots)
roots = Bisection(a, b, Nmax, eps)
plt.figure()
plt.plot(range(len(roots)), roots, marker='o')
plt.xlabel("Iteration")
plt.ylabel("Root estimate")
plt.title("Convergence of Bisection Method")
plt.grid(True)
plt.show()

print(roots)
print(f'root of the f(x) is {roots[-1]}')   #1.0296630859375
print(f'value of f(x) at the root: {f(roots[-1])}')   #0.0005522177242860415
print(f'is value of the fn at the root less then eps: {f(roots[-1]) < eps}') #True
```
