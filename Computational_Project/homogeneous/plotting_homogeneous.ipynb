```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
plt.rcParams["figure.figsize"] = (10,6)
plt.rcParams["font.size"] = 12
```

```python
# homogeneous_case_1: underdamped (c=0.5)
# homogeneous_case_2: intermediate decay (c=2.45)
# homogeneous_case_3: fast decay ((c=4.9))
```

```python
# ==========================================================
# READ FILE
# ==========================================================
file_name = ["homogeneous_case_1.csv", "homogeneous_case_2.csv", "homogeneous_case_3.csv"] 
df1 = pd.read_csv(file_name[0])
df2 = pd.read_csv(file_name[1])
df3 = pd.read_csv(file_name[2])
```

```python
t = df1["time"]

x1 = df1["x"]
x2 = df2["x"]
x3 = df3["x"]

v1 = df1["v"]
v2 = df2["v"]
v3 = df3["v"]

a1 = df1["a"]
a2 = df2["a"]
a3 = df3["a"]

f_spring1 = df1["f_spring"]
f_damp1 = df1["f_damp"]
f_total1 = df1["f_total"]
f_spring2 = df2["f_spring"]
f_damp2 = df2["f_damp"]
f_total2 = df2["f_total"]
f_spring3 = df3["f_spring"]
f_damp3 = df3["f_damp"]
f_total3 = df3["f_total"]

ke1 = df1["ke"]
pe1 = df1["pe"]
te1 = df1["te"]
ke2 = df2["ke"]
pe2 = df2["pe"]
te2 = df2["te"]
ke3 = df3["ke"]
pe3 = df3["pe"]
te3 = df3["te"]
```

```python
# ==========================================================
# PLOT DISPLACEMENT vs TIME
# ==========================================================
plt.figure()
plt.plot(t, x1, color='b', label='Underdamped')
plt.plot(t, x2, color='g', label='Critically damped')
plt.plot(t, x3, color='r', label='Overdamped')
# plt.ylim([-1.1, 1.1])
plt.xlabel(r"Time $t$")
plt.ylabel(r"Position $x(t)$")
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend()
plt.tight_layout()
plt.savefig("position.png", dpi=600, bbox_inches="tight")
plt.show()
```

```python
# ==========================================================
# PLOT VELOCITY vs TIME
# ==========================================================
plt.figure()
plt.plot(t, v1, color='b', label='Underdamped')
plt.plot(t, v2, color='g', label='Critically damped')
plt.plot(t, v3, color='r', label='Overdamped')
# plt.ylim([-2.1, 2.1])
plt.xlabel(r"Time $t$")
plt.ylabel(r"Velocity $v(t)$")
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend()
plt.tight_layout()
plt.savefig("velocity.png", dpi=600, bbox_inches="tight")
plt.show()
```

```python
# ==========================================================
# PLOT ACCELERATION vs TIME
# ==========================================================
plt.figure()
plt.plot(t, a1, color='b', label='Underdamped')
plt.plot(t, a2, color='g', label='Critically damped')
plt.plot(t, a3, color='r', label='Overdamped')
plt.xlabel(r"Time $t$")
plt.ylabel(r"Acceleration $a(t)$")
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend()
plt.tight_layout()
plt.savefig("accn.png", dpi=600, bbox_inches="tight")
plt.show()
```

```python
# ==========================================================
# PLOT FORCE vs TIME (UNDERDAMPED)
# ==========================================================
plt.figure()
plt.plot(t, f_spring1, color='b', label='Spring $F_s$')
plt.plot(t, f_damp1, color='g', label='Damping $F_d$')
plt.plot(t, f_total1, color='r', label='Total $F$')
plt.xlabel(r"Time $t$")
plt.ylabel(r"Force $F(t)$")
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend()
plt.title("Force vs Time | Underdamped Case")
plt.tight_layout()
plt.savefig("force_under.png", dpi=600, bbox_inches="tight")
plt.show()
```

```python
# ==========================================================
# PLOT FORCE vs TIME (CRITICALLY DAMPED)
# ==========================================================
plt.figure()
plt.plot(t, f_spring2, color='b', label='Spring $F_s$')
plt.plot(t, f_damp2, color='g', label='Damping $F_d$')
plt.plot(t, f_total2, color='r', label='Total $F$')
plt.xlabel(r"Time $t$")
plt.ylabel(r"Force $F(t)$")
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend()
plt.title("Force vs Time | Critically Damped Case")
plt.tight_layout()
plt.savefig("force_crit.png", dpi=600, bbox_inches="tight")
plt.show()
```

```python
# ==========================================================
# PLOT FORCE vs TIME (OVERDAMPED)
# ==========================================================
plt.figure()
plt.plot(t, f_spring3, color='b', label='Spring $F_s$')
plt.plot(t, f_damp3, color='g', label='Damping $F_d$')
plt.plot(t, f_total3, color='r', label='Total $F$')
plt.xlabel(r"Time $t$")
plt.ylabel(r"Force $F(t)$")
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend()
plt.title("Force vs Time | Overdamped Case")
plt.tight_layout()
plt.savefig("force_over.png", dpi=600, bbox_inches="tight")
plt.show()
```

```python
# ==========================================================
# PLOT ENERGY vs TIME (UNDERDAMPED)
# ==========================================================
plt.figure()
plt.plot(t, ke1, color='r', label='KE')
plt.plot(t, pe1, color='b', label='PE')
plt.plot(t, te1, color='k', alpha=0.7, label='$Total$')
plt.xlabel(r"Time $t$")
plt.ylabel(r"Energy")
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend()
plt.title("Energy vs Time | Underdamped Case")
plt.tight_layout()
plt.savefig("energy_under.png", dpi=600, bbox_inches="tight")
plt.show()
```

```python
# ==========================================================
# PLOT ENERGY vs TIME (CRITICALLY DAMPED)
# ==========================================================
plt.figure()
plt.plot(t, ke2, color='r', label='KE')
plt.plot(t, pe2, color='b', label='PE')
plt.plot(t, te2, color='k', alpha=0.7, label='$Total$')
plt.xlabel(r"Time $t$")
plt.ylabel(r"Energy")
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend()
plt.title("Energy vs Time | Critically Damped Case")
plt.tight_layout()
plt.savefig("energy_crit.png", dpi=600, bbox_inches="tight")
plt.show()
```

```python
# ==========================================================
# PLOT ENERGY vs TIME (OVERDAMPED)
# ==========================================================
plt.figure()
plt.plot(t, ke3, color='r', label='KE')
plt.plot(t, pe3, color='b', label='PE')
plt.plot(t, te3, color='k', alpha=0.7, label='$Total$')
plt.xlabel(r"Time $t$")
plt.ylabel(r"Energy")
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend()
plt.title("Energy vs Time | Overdamped Case")
plt.tight_layout()
plt.savefig("energy_over.png", dpi=600, bbox_inches="tight")
plt.show()
```

```python
# ==========================================================
# PLOT PHASE SPACE (UNDERDAMPED)
# ==========================================================
plt.figure()
plt.plot(x1, v1, lw=1, color='b')
plt.ylabel(r"Velocity $v(t)$")
plt.xlabel(r"Position $x(t)$")
plt.grid(True, linestyle='--', alpha=0.7)
plt.title("Phase Space | Underdamped Case")
plt.tight_layout()
plt.savefig("phase_under.png", dpi=600, bbox_inches="tight")
plt.show()
```

```python
# ==========================================================
# PLOT PHASE SPACE (CRITICALLY DAMPED)
# ==========================================================
plt.figure()
plt.plot(x2, v2, lw=1, color='g')
plt.ylabel(r"Velocity $v(t)$")
plt.xlabel(r"Position $x(t)$")
plt.grid(True, linestyle='--', alpha=0.7)
plt.title("Phase Space | Critically Damped Case")
plt.tight_layout()
plt.savefig("phase_crit.png", dpi=600, bbox_inches="tight")
plt.show()
```

```python
# ==========================================================
# PLOT PHASE SPACE (OVERDAMPED)
# ==========================================================
plt.figure()
plt.plot(x3, v3, lw=1, color='r')
plt.ylabel(r"Velocity $v(t)$")
plt.xlabel(r"Position $x(t)$")
plt.grid(True, linestyle='--', alpha=0.7)
plt.title("Phase Space | Overdamped Case")
plt.tight_layout()
plt.savefig("phase_over.png", dpi=600, bbox_inches="tight")
plt.show()
```
