```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.ndimage import gaussian_filter1d
from scipy.signal import savgol_filter
```

```python
plt.rcParams["figure.figsize"] = (10,6)
plt.rcParams["font.size"] = 12
```

```python
# deterministic_exp_case_1: slow decay
# deterministic_exp_case_2: intermediate decay
# deterministic_exp_case_3: fast decay
```

```python
# ==========================================================
# READ FILE
# ==========================================================
file_name = "deterministic_exp_case_3.csv"
df = pd.read_csv(file_name)
```

```python
t = df["time"]
x = df["x"]
v = df["v"]
# x = np.sqrt(df["var_x"]   + df["mean_x"]**2) #rms
# v = np.sqrt(df["var_v"]   + df["mean_v"]**2) #rms
a = df["a"]
# f_noise = df["mean_eta"]
f_spring = df["f_spring"]
f_damp = df["f_damp"]
f_drive = df["f_drive"]
f_total = df["f_total"]
ke = df["ke"]
pe = df["pe"]
te = df["te"]
```

```python
# ==========================================================
# PLOT DISPLACEMENT vs TIME
# ==========================================================
plt.figure(figsize=(10,3))
plt.plot(t, x, lw=1.5, color='b')
# plt.ylim([-1.1, 1.1])
plt.xlabel(r"Time $t$")
plt.ylabel(r"Position $x(t)$")
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.savefig("position_exp.png", dpi=600, bbox_inches="tight")
plt.show()
```

```python
# ==========================================================
# PLOT VELOCITY vs TIME
# ==========================================================
plt.figure(figsize=(10,3))
plt.plot(t, v, lw=1.5, color='r')
# plt.ylim([-2.1, 2.1])
plt.xlabel(r"Time $t$")
plt.ylabel(r"Velocity $v(t)$")
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.savefig("velocity_exp.png", dpi=600, bbox_inches="tight")
plt.show()
```

```python
# ==========================================================
# PLOT ACCELERATION vs TIME
# ==========================================================
# a_smooth = savgol_filter(a, 501, 3)
plt.figure(figsize=(10,3))
plt.plot(t, a, lw=1.5, color='k')
# plt.plot(t, a, lw=1, alpha=0.5)
# plt.ylim([-5.1, 5.1])
plt.xlabel(r"Time $t$")
plt.ylabel(r"Acceleration $a(t)$")
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.savefig("accn_exp.png", dpi=600, bbox_inches="tight")
plt.show()
```

```python
# ==========================================================
# PLOT FORCE vs TIME
# ==========================================================
# f_noise_smooth = gaussian_filter1d(f_noise, sigma=10)
# f_noise_smooth = savgol_filter(f_noise, 501, 3)
plt.figure()
plt.plot(t, f_spring, color='b', label='Spring $F_s$')
plt.plot(t, f_damp, color='g', label='Damping $F_d$')
plt.plot(t, f_drive, color='r', label='Driving $F(t)$')
# plt.plot(t, f_damp+f_spring+f_drive, color='k', label='Total $F=F_s+F_d+F_{OU}$')
# plt.plot(t, f_total, lw=0.5, color='k', alpha=0.5)
# plt.ylim([-5.1, 5.1])
plt.xlabel(r"Time $t$")
plt.ylabel(r"Force $F(t)$")
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend()
plt.tight_layout()
plt.savefig("force_exp.png", dpi=600, bbox_inches="tight")
plt.show()
```

```python
# ==========================================================
# PLOT ENERGY vs TIME
# ==========================================================
plt.figure()
plt.plot(t, ke, color='r', label='KE')
plt.plot(t, pe, color='b', label='PE')
plt.plot(t, te, color='k', alpha=0.7, label='$Total$')
# plt.ylim([-5.1, 5.1])
plt.xlabel(r"Time $t$")
plt.ylabel(r"Energy")
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend()
plt.tight_layout()
plt.savefig("energy_exp.png", dpi=600, bbox_inches="tight")
plt.show()
```

```python
# ==========================================================
# PLOT PHASE SPACE
# ==========================================================
plt.figure()
plt.plot(x, v, lw=1, color='r')
# plt.ylim([-5.1, 5.1])
plt.ylabel(r"Velocity $v(t)$")
plt.xlabel(r"Position $x(t)$")
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.savefig("phase_exp.png", dpi=600, bbox_inches="tight")
plt.show()
```
