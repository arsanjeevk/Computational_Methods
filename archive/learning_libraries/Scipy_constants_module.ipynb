## scipy.constants

The scipy.constants module is SciPy’s library for scientific constants, unit conversions, and physical quantity utilities.

It is built to provide accurate, up-to-date CODATA (Committee on Data for Science and Technology) values.

```python
from scipy import constants as const   #Import constants module
import numpy as np
```

## Most commonly used constants

```python
const.c        # Speed of light (m/s)
const.h        # Planck constant (J·s)
const.hbar     # Planck constant / (2π)
const.G        # Gravitational constant
const.e        # Elementary charge (C)
const.N_A      # Avogadro's number
const.k        # Boltzmann constant (J/K)
const.R        # Gas constant (J/(mol·K))
const.g        # Standard gravity (m/s^2)
```

## CODATA — Constants With Uncertainty

All fundamental constants include uncertainty + units.

```python
from scipy.constants import physical_constants

print(physical_constants['Planck constant'])
print(physical_constants['Boltzmann constant'])
print(physical_constants['Avogadro constant'])
print(physical_constants['Stefan-Boltzmann constant'])
```

## --> Important PHYSICS constants

```python
# ============================
# FUNDAMENTAL CONSTANTS
# ============================

const.c           # Speed of light in vacuum (m/s)
const.h           # Planck constant (J·s)
const.hbar        # Reduced Planck constant ħ = h/(2π) (J·s)
const.G           # Newtonian gravitational constant (m^3·kg^-1·s^-2)
const.e           # Elementary charge (C)
const.N_A         # Avogadro's number (1/mol)
const.k           # Boltzmann constant (J/K)
const.R           # Gas constant (J/(mol·K))
const.g           # Standard gravity (m/s^2)
const.mu_0        # Vacuum permeability (N/A^2)
const.epsilon_0   # Vacuum permittivity (F/m)
const.alpha       # Fine-structure constant (dimensionless)
const.physical_constants   # Access to all CODATA constants


# ============================
# PARTICLE PHYSICS CONSTANTS
# ============================

const.m_e         # Electron mass (kg)
const.m_p         # Proton mass (kg)
const.m_n         # Neutron mass (kg)
const.m_mu        # Muon mass (kg)
const.m_tau       # Tau mass (kg)

const.electron_mass        # Electron mass (same as m_e)
const.proton_mass          # Proton mass (same as m_p)
const.neutron_mass         # Neutron mass (same as m_n)

const.atomic_mass          # Atomic mass constant (1 u in kg)
const.physical_constants['Rydberg constant']       # Rydberg constant
const.physical_constants['Bohr radius']            # Bohr radius (m)
const.physical_constants['Bohr magneton']          # Bohr magneton (J/T)
const.physical_constants['nuclear magneton']       # Nuclear magneton (J/T)
const.physical_constants['electron g factor']      # Electron g-factor


# ============================
# ELECTROMAGNETIC CONSTANTS
# ============================

const.mu_0                 # Vacuum permeability (N/A²)
const.epsilon_0            # Vacuum permittivity (F/m)
const.c                    # Speed of light (m/s)
const.Z0                   # Impedance of free space (ohms)
const.sigma                # Stefan–Boltzmann constant (W/m²·K⁴)
const.physical_constants['Bohr magneton']          # Bohr magneton
const.physical_constants['nuclear magneton']       # Nuclear magneton


# ============================
# QUANTUM CONSTANTS
# ============================

const.h                    # Planck constant (J·s)
const.hbar                 # Reduced Planck constant ħ
const.physical_constants['Rydberg constant']       # R∞
const.physical_constants['Bohr radius']            # a₀
const.physical_constants['Hartree energy']         # Hartree energy (J)
const.physical_constants['electron volt']          # 1 eV in Joules


# ============================
# THERMODYNAMICS CONSTANTS
# ============================

const.k                    # Boltzmann constant (J/K)
const.R                    # Universal gas constant (J/(mol·K))
const.N_A                  # Avogadro constant
const.sigma                # Stefan–Boltzmann constant (W/m²·K⁴)
const.physical_constants['Wien displacement constant']   # Wien constant


# ============================
# ASTRONOMY CONSTANTS
# ============================

const.au                   # Astronomical unit (m)
const.light_year           # Light-year (m)
const.parsec               # Parsec (m)

const.physical_constants['solar mass']              # Solar mass (kg)
const.physical_constants['astronomical unit']       # AU (same as const.au)


# ============================
# NUCLEAR CONSTANTS
# ============================

const.physical_constants['atomic mass constant']       # 1 u in kg
const.physical_constants['neutron mass']               # Neutron mass
const.physical_constants['proton mass']                # Proton mass
const.physical_constants['fine-structure constant']    # α (dimensionless)


# ============================
# RADIATION CONSTANTS
# ============================

const.sigma                # Stefan-Boltzmann constant (W/m²·K⁴)
const.physical_constants['radiation constant']   # a = 4σ/c (J/m³·K⁴)

```

## --> Important mathematical constants

```python
const.pi        # Pi (π): Ratio of circumference to diameter
const.tau       # Tau (τ = 2π): One full turn in radians
const.e         # Euler's number (e): Base of natural logarithm
const.euler     # Euler–Mascheroni constant (γ)
const.golden    # Golden ratio (φ = (1 + √5) / 2)
const.degree    # 1 degree in radians (π/180)
const.arcmin    # 1 arcminute in radians (1/60 degree)
const.arcsec    # 1 arcsecond in radians (1/3600 degree)

const.rad2deg   # Radians → Degrees conversion factor
const.deg2rad   # Degrees → Radians conversion
print(const.e)
```
