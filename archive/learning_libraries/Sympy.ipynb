```python
from sympy import *
init_printing()

x, y, z = symbols('x y z')
a = Symbol('a', real=True)
n = Symbol('n', integer=True)
expr = x**2 + 2*x + 1
expr.subs(x, 2)
expr.subs(x, 2).evalf()

expand((x+1)**3)

factor(x**2 + 2*x + 1)

simplify((x**2-1)/(x-1))

collect(x**2 + 2*x + x**2, x)

cancel((x**2-1)/(x-1))

#Solving equations
solve(x**2 - 4, x)

solve([x+y-2, x-y-1], [x,y])

solve(a*x + 5, x)

nsolve(x**2-2, x, 1)
expr = x ** 2
expr.subs(x, 2)
expr.subs(x, 2).evalf()



# Calculas

diff(x**3, x)
diff(sin(x)*exp(x), x)
diff(x**3, x, 2)   # second derivative


integrate(x**2, x)
integrate(sin(x), (x,0,pi))
integrate(sin(x)/x, (x, 0, oo))

#Limit
limit(sin(x)/x, x, 0)

limit(sin(x)/x, x, 0)
limit(1/x, x, oo)

#Taylor series
series(sin(x), x, 0, 6)

#Differential equation
y = Function('y')
dsolve(Eq(y(x).diff(x,2) + y(x), 0))




#Matrices and Linear Algebra

A = Matrix([[1,2],[3,4]])

A.T      # transpose
A.det()
A.inv()
A.trace()
A.eigenvals()
A.eigenvects()

A.LUsolve(Matrix([1,2])) #Solve linear system

#identity and zero
eye(3)
zeros(2,2)



#Physics style symbolic tools

pi, E, I, oo  # pie, e, iota, infinty


conjugate(1+I)
re(3+4*I)
im(3+4*I)

trigsimp(sin(x)**2 + cos(x)**2) #simplify trignomatery


#laplace transforms
# laplace_transform(sin(x), x, s)
# inverse_laplace_transform(1/(s**2+1), s, x)



#Fouries Transforms
k = Symbol('k', integer=True)
fourier_transform(exp(-x**2), x, k)

#pretty printing (nice output)
pprint(expr)
latex(expr)


```
