#Linear Congruential Generator (LCG)

core formula
$$X_{n+1}  = reminder(\frac{(a X_n + c)} {m})$$
where
| Symbol    | Meaning        |
| --------- | -------------- |
| $X_n$     | current number |
| $X_{n+1}$ | next number    |
| a      | multiplier     |
| c       | increment      |
| m       | modulus        |
| $X_0$     | seed           |


```python
import numpy as np
import matplotlib.pyplot as plt

def lcg(seed,a,c,m,n):
  numbers = []
  for i in range(500):
    nums = (a * seed + c) % m
    numbers.append(nums)
    seed = nums
    i += 1
  return numbers/np.max(numbers)

numbers = lcg(7,5,3,16,500)
print(numbers)
plt.hist(numbers)
plt.show()
```

#Uniform Random numbers Vs Gaussian Random numbers

| Feature           | Uniform          | Gaussian              |
| ----------------- | ---------------- | --------------------- |
| Shape             | flat             | bell                  |
| Most likely value | all equal        | mean                  |
| Extremes          | common           | rare                  |
| Used for          | basic randomness | natural noise/physics |
| Symbol            | U(a,b)           | N(μ,σ²)               |


#### Gaussian = “nature’s default randomness:



```python
import matplotlib.pyplot as plt
import numpy as np
x_uniform = np.random.rand(500000)
x_gaussian = np.random.randn(500000)

fig, ax = plt.subplots(1, 2, figsize=(15,5))
ax[0].hist(x_uniform, bins=50)

ax[1].hist(x_gaussian, bins=50)
plt.show()
```

#Box Muller Transformation

```python
import numpy as np
import matplotlib.pyplot as plt

#Generating two uniform distributions
u1 = np.random.rand(100000)
u2 = np.random.rand(100000)

# r → distance from center (points in 2D)
r = np.sqrt( -2 * np.log(u1))

#x and y are Gaussian random variables generated from uniform random numbers
x = r * np.cos(2 * np.pi * u2)
y = r * np.sin(2 * np.pi * u2)

fig, ax = plt.subplots(1,5, figsize=(25,5))
ax[0].hist(u1)
ax[1].hist(u2)
ax[2].scatter(x,y)
ax[3].hist(x)
ax[4].hist(y)
plt.show()
```

#Make Double bell curve

```python
import numpy as np
import matplotlib.pyplot as plt

N = 10000
z = []

for i in range(N):
    if np.random.rand() < 0.5:          # coin flip
        z.append(np.random.normal(-2, 0.5))  # left bell
    else:
        z.append(np.random.normal(2, 0.5))   # right bell

plt.hist(z, bins=80, density=True)
plt.show()

```
