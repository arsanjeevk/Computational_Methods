```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
  return 2 * np.cos(x) - x

def df(x):
  return -2 * np.sin(x) - 1


#Newto raphson method
Nmax = 100
eps = 1e-3
x_0 = 6

def newton_raphson_method(x_0, Nmax, eps):
    roots = []
    x = x_0
    for it in range(Nmax):
        roots.append(x)

        x = x - f(x) / df(x)

        if abs(f(x)) < eps:
            print("Converged in", it+1, "iterations")
            break

    return np.array(roots)
roots = newton_raphson_method(x_0, Nmax, eps)
print(roots[-1])
print(f(roots[-1]))
plt.figure()
plt.plot(range(len(roots)), roots, marker='o')
plt.xlabel("Iteration")
plt.ylabel("Root estimate")
plt.title("Convergence of Newton-Raphson Method")
plt.grid(True)
```

```python
import numpy as np

def f(x):
    return 2 * np.cos(x) - x

def df(x):
    return -2 * np.sin(x) - 1


def newton_raphson(x0, N_max=100, eps=1e-6):
    x = x0
    history = []   # store x_n
    deltas = []    # store Δx_n

    for n in range(N_max):
        x_new = x - f(x) / df(x)     # Newton step
        delta_x = x_new - x          # Δx = x_{n+1} − x_n

        history.append(x)
        deltas.append(delta_x)

        # convergence check (both criteria)
        if abs(f(x_new)) < eps or abs(delta_x) < eps:
            print(f"Converged in {n+1} iterations")
            history.append(x_new)
            deltas.append(0.0)
            break

        x = x_new

    return np.array(history), np.array(deltas)

x0 = 6
x_vals, delta_x_vals = newton_raphson(x0, N_max=100, eps=1e-6)

print("Final root approximation:", x_vals[-1])
print("Function value:", f(x_vals[-1]))
print("Last Δx:", delta_x_vals[-1])
```
